### import and configure
**import** the **matplotlib** module

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

### draw a graph
use **plot** and **show** to create a graph

```python
xs = [1, 2, 5]
ys = [4, 7, 5]

plt.plot(xs, ys)
plt.show()
```

### create data for graphs
plot can be called multiple times to overlay graphs

```python
from random import uniform

min_deviation = -10
max_deviation = 10

xs = [x / 10 for x in range(0, 100)]
ys = [x * x for x in xs]
ys_deviated = [x * x + uniform(min_deviation, max_deviation) for x in xs]

plt.plot(xs, ys)
plt.plot(xs, ys_deviated)
plt.show()
```
